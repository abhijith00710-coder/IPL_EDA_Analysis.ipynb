```python
import pandas as pd

matches=pd.read_csv('/content/matches.csv')
deliveries=pd.read_csv('/content/deliveries.csv')

# view first rows
matches.head()
```

```python
# check type
matches.info()  # there is no type change
```

```python
deliveries.info()
```

```python
matches.describe(),deliveries.describe()  # statistical summary
```

```python
#missing value checking

matches.isnull().sum(),deliveries.isnull().sum()
```

```python
matches.drop('method',axis=1,inplace=True)
matches.info() # removed methode column because it contains 90% of missing value
```

```python
# checking duplicate values

matches.duplicated().sum(),deliveries.duplicated().sum() # there is no duplicate values
```

```python
# fill missing values

# in deliveries the missing value may be treated as 'zero'

# in matches the missing values are of the form descriptive so replace it by 'unknown' and missing values are of the form float replaced by 'zero'

matches['result_margin'].fillna(0,inplace=True)
matches['target_runs'].fillna(0,inplace=True)
matches['target_overs'].fillna(0,inplace=True)
matches.fillna('unknown',inplace=True)

deliveries.fillna(0,inplace=True)
```

```python
deliveries.isnull().sum() , matches.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Question 1: Which team won maximum matches?
wins=matches['winner'].value_counts()
wins
```

```python
wins.plot(kind='bar')
plt.xticks(rotation=90)
plt.ylabel('Number of wins')
plt.title("total wins by team")
plt.show()

# we can clearly seen that MI,CSk,KKR are dominating throughout the entire ipl season
```

```python
# Question 2: Toss winner vs Match winner

toss_match=matches['toss_winner']==matches['winner']
toss_match.value_counts()
```

```python
toss_percentage=toss_match.mean()*100
toss_percentage
```

```python
sns.countplot(x=toss_match)
plt.title("Toss winner won match?")
plt.show()

# toss winning does not guarantee a victory , there is still 50-50 chance
```

```python
venue=matches['venue'].value_counts()

venue.head(10) # eden gardens top on the list ,it shows that highest number of matches played in this venue
```

```python
# Top 10 run scorers

top_runs=deliveries.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False)

top_runs.head(10)
```

```python
top_runs.head(10).plot(
kind='barh')

plt.title("Top Run Scorers")

plt.show()
```

```python
# top 10 wicket takers

top_wicket=deliveries.groupby('bowler')['is_wicket'].sum().sort_values(ascending=False)
top_wicket.head(10)
```

```python
top_wicket.head(10).plot(kind='bar')
plt.title('Top 10 Wicket Takers')
plt.show()
```

```python
# strike rate= (Runs/balls)*100

runs=deliveries.groupby(
'batter')['batsman_runs'].sum()

balls=deliveries.groupby(
'batter')['ball'].count()

strike=(runs/balls)*100

strike.sort_values(
ascending=False).head(10)
```

```python
# Economy = (Runs conceded / overs)

runs_conceded=deliveries.groupby('bowler')['total_runs'].sum()
balls=deliveries.groupby('bowler')['ball'].count()

overs=balls/6

ecmy=(runs_conceded/overs)
ecmy.sort_values().head(10)
```

```python
ecmy.head(10).plot(kind='barh')
plt.show()
```
